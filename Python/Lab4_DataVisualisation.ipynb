```python
plt.plot(4,3,'.')
plt.show()
```

#### PLOT NUMBER OF POINTS

```python
x = np.array([2,4,6,8,10,12,14,16])
y = x/2

plt.figure(figsize=(10,5))
plt.scatter(x,y,c='green')
plt.show()
```

#### LABEL THE AXIS

```python
plt.figure(figsize=(10,5))
plt.plot([1,2,3,4],c='green')
plt.xlabel('horizontal')
plt.ylabel('vertical')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot([1,2,3,4],c='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

#### MULTIPLE PLOTS WITH SUBPLOTS

```python
classes = ['John','Sanjay','Ellen','Sandra','Nancy','Shubham','Philip']
scores = [10,15,20,30,40,50,100]

plt.figure(figsize=(20,7))
plt.subplot(131)
plt.bar(classes, scores)

plt.subplot(132)
plt.scatter(classes,scores)

plt.subplot(133)
plt.plot(classes,scores)

plt.suptitle('Categorical Plotting')
plt.show()
```

#### SEABORN

```python
ourdata = pd.read_csv('Pokemon.csv')

ourdata.head()
```

```python
sns.lmplot(x='Attack',y='Defense',data=ourdata)
plt.show()
```

```python
sns.lmplot(x='Attack',y='Defense',data=ourdata,fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='Attack',y='Defense',data=ourdata,fit_reg=False,hue='Legendary')
plt.show()
```

```python
ourdata['Legendary'].value_counts()
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])

x = np.arange(10)
a1.plot(x,np.exp(x),'b')

a1.set_title('range of numbers')

plt.ylim(0,10000)
plt.xlim(0,10)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
